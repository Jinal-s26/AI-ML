# Data Cleaning

1. Analyse the data and identify which columns are not relevant for house rent prediction task. Drop those columns from the dataframes.

```python
test1 = test.drop(['Posted On', 'Tenant Preferred', 'Point of Contact'], axis=1)
test1
```

2. Check for missing values and logically impute the dataset.

```python
print(test1.isnull())
print(test1.isnull().sum())
```

No imputation required as no null value is there in the dataset.

3. Identify any categorical valued columns (non-numeric) and convert them to numeric.

```python
test1["Area Type"] = test1["Area Type"].map({"Super Area": 1, 
                                           "Carpet Area": 2, 
                                           "Built Area": 3})
test1["City"] = test1["City"].map({"Mumbai": 0, "Chennai": 1, 
                                 "Bangalore": 2, "Hyderabad": 3, 
                                 "Delhi": 4, "Kolkata": 5})
test1
```

# Exploratory Analysis (On training set):

1. Plot the house rents against the dependent variable of “size”. See if there is a uniform linear trend between the dependent and independent variables. Make accurate axis and legend. Save the plot in a png file.

```python
train.plot(x="Rent", y="Size", kind="line", figsize=(10, 10))
plt.title('House rent against size')
plt.ylabel('Size of Houses')
plt.xlabel("Rent of Houses")
```

3. Find average rent prices in different cities and report which city has the highest average rent.

```python
df = train.groupby('City', as_index=False)['Rent'].mean()
print(df)
```

```python
plt.bar(df["City"], df["Rent"], color = ["#4188ff", "#649eff", "#78aaff", "#92bbff", "#accbff"])
```

Highest Average Rent is in Mumbai.
