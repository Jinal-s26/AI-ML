```python
iS = [[3,7,6],[5,1,2],[4,0,8]]
gs = [[5,3,6], [7,0,2], [4,1,8]]

def print_mat(mat):
    for i in range(len(mat)):
        for j in range(len(mat)):
            print(mat[i][j], end =" ")
        print(" ")
        
def search(mat, value):
    for i in range(3):
        for j in range(3):
            if mat[i][j] == value:
                return i, j

# def swap(mat, row, col, r, c):
    
 

    
md = [0,0,0,0,0,0,0,0]

for i in range(len(md)):
    rpos_is, cpos_is = search(iS,i+1)
    rpos_gs, cpos_gs = search(gs,i+1)
    md[i] = abs(rpos_is-rpos_gs) + abs(cpos_is-cpos_gs)

print(md)
cur_md = sum(md)
row, col = search(iS, 0)
dir_pos = [[-1,0], [1,0], [0,-1], [0,1]]
#           u        d       l       r
step = 1
while(iS!=gs):
    r = 0
    c = 0
    if row+dir_pos[0][0]>=0:
        num = iS[row+dir_pos[0][0]][col+dir_pos[0][1]]
        rgu, cgu = search(gs, num)
        up = cur_md + abs(rgu - row) + abs(cgu - col) - md[num-1]
    else:
        up = 999999
        
    if row+dir_pos[1][0]<3:
        num = iS[row+dir_pos[1][0]][col+dir_pos[1][1]]
        rgd, cgd = search(gs, num)
        down = cur_md + abs(rgd - row) + abs(cgd - col)-md[num-1]
    else:
        down = 999999
        
    if col+dir_pos[2][1]>=0:
        num = iS[row+dir_pos[2][0]][col+dir_pos[2][1]]
        rgl, cgl = search(gs, num)
        left = cur_md + abs(rgl - row) + abs(cgl - col) - md[num-1]
    else:
        left = 999999
        
    if col+dir_pos[3][1]<3:
        num = iS[row+dir_pos[3][0]][col+dir_pos[3][1]]
        rgr, cgr = search(gs, num)
        right = cur_md + abs(rgr - row) + abs(cgr - col) - md[num-1]
    else:
        right = 999999
    
    if up<down and up<left and up<right:
        curd_md = up
        r,c = (row+dir_pos[0][0]), (col+dir_pos[0][1]) 
        
    if down<up and down<left and down<right:
        curd_md = down
        r,c = row+dir_pos[1][0], col+dir_pos[1][1]

    if left<down and left<up and left<right:
        curd_md = left
        r,c = row+dir_pos[2][0], col+dir_pos[2][1]
    
    if right<down and right<left and right<up:
        curd_md = right
        r,c = row+dir_pos[3][0], col+dir_pos[3][1]
        
    iS[row][col] = iS[r][c]
    iS[r][c] = 0
#     print(iS[r][c],iS[row][col])
    rgr, cgr = search(gs,iS[row][col])
    md[iS[row][col]-1] = abs(rgr - row) + abs(cgr - col)
    row, col = search(iS, 0)
    print("Step ", step)
    step = step + 1
    print(md)
    print_mat(iS)
    print(" ")
```
